CÓDIGO PARA RENOMBRAR LAS IMÁGENES DE ESPECTROGRAMAS

Este código se utilizó para renombrar las imágenes de espectrogramas de los ruidos de motores elevadores proporcionadas por el laboratorio de ruido de Bosch México planta Toluca, Edo. de México.
Las imágenes se encuentran divididas en dos carpetas según sean una muestra aceptable o no aceptable de ruido de un motor y han sido etiquetadas con un ID y un ok o nok. El conjunto de datos etiquetado está almacenado en la carpeta de google drive del equipo.

```python
import os
from google.colab import drive
drive.mount('/content/drive')
```

```python
folder_input_nok = '/content/drive/My Drive/Proyecto Integrador Bosch/processed_images/nok_500'
folder_output_nok = '/content/drive/My Drive/Proyecto Integrador Bosch/processed_images/NOK_etiquetadas_500'

images = os.listdir(folder_input_nok)
```

```python
for i, image in enumerate(images):
  original_path = os.path.join(folder_input_nok, image) #ruta completa de cada archivo

  new_label = f'nok.({i+1})' + os.path.splitext(image)[1]

  new_path = os.path.join(folder_output_nok, new_label)

  os.rename(original_path,new_path)

  print(f'Renombrado {image} a {new_label} y movido a NOK_etiquetadas_500')

drive.flush_and_unmount()
print('Etiquetado completado')
```
